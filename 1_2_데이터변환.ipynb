## sklearn

```python
import pandas as pd
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/AnYeHyeon/DataScience_2023S/master/data/auto.csv')
df
```

```python
# 객체생성
scaler = StandardScaler()
output = scaler.fit_transform(df) #numpy array 형태로 변환시켜서 생성
output
```

```python
pd.DataFrame(output, columns = df.columns)
```

```python
scaler = MinMaxScaler()
output = scaler.fit_transform(df)
pd.DataFrame(output, columns = df.columns)
```

# Polynormial Term 추가하기
- 선형회귀 모형의 한계를 깨기 위한 방법
- 비선형적인 관계를 표현하는 방법
    - (회귀) 다차항 넣기
    - (분류) 변수간의 상호작용
- sklean에서 제공하는 함수로 손쉽게 해결 가능

```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/AnYeHyeon/DataScience_2023S/master/data/classification_data.csv")
df
```

```python
# 독립변수 추리기
x = df[['X1', 'X2']]
x
```

```python
from sklearn import preprocessing

from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(3, include_bias=False, interaction_only=False) # 몇차식, 상수항(편향), 상호작용 
output = poly.fit_transform(x)
```

```python
c = poly.get_feature_names_out(x.columns)
```

```python
pd.DataFrame(output, columns = c)
```
