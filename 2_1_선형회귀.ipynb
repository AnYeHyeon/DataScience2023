# 가상데이터 만들기

```python
import numpy as np
```

```python
# 맞춰야 하는 정답
true_w = 2
true_b = -3
```

```python
# 평균이 0이고 표준편차가 1인 정규분포로부터
# 50개의 데이터를 무작위로 추출
X = np.random.normal(1, size=50)
X
```

```python
# y는 다음과 같은 관계식에 의해 얻어졌다. (가정)
y = true_w*X + true_b # y는 2X-3
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y)
```

```python
# noise
noise = np.random.normal(0.1, size=50)
noise
```

```python
y = y + noise
```

```python
y
```

```python
plt.scatter(X, y)
```
