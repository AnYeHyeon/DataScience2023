# 데이터 불러오기

```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/AnYeHyeon/DataScience_2023S/master/data/titanic_processed.csv')
df
```

```python
import numpy as np
```

```python
np.sum(df['survived'])
```

# 학습

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
```

# 데이터 분리 (종속, 독립)

```python
X = df[df.columns[:-1]]
X
```

```python
y = df['survived']
```

```python
X, y
```

```python
# 학습
reg.fit(X, y)
```

# 정확도 측정

```python
y_pred = reg.predict(X)
y_pred
```

```python
import numpy as np
np.sum(y_pred == y)/y.count()
```

# 계수의 방향과 크기를 가지고 모델을 분석

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (16, 8))
plt.bar(X.columns, reg.coef_[0])
```

## 데이터의 표준화(정규화)

```python
x_norm = (X-X.mean())/X.std()
```

```python
reg.fit(x_norm, y)
```

```python
plt.figure(figsize = (16, 8))
plt.bar(X.columns, reg.coef_[0])
```
