# 변수의 표준화
- 표준화 하기전
    - 변수간에 서로 다른 Scale, unit
- 특히 거리 기반의 머신러닝 방법
    - standardize (표준화): 변수가 평균이 0, 표준편차가 1이 되도록 변환
    - min-max scaling: 변수 내에서 가장 큰 값1, 작은값 0

```python
df = pd.read_csv('https://raw.githubusercontent.com/AnYeHyeon/DataScience_2023S/master/data/auto.csv')
df
```

## Standardize

```python
# 공식: (data - mean) / (std)
df.mean() 
```

```python
df - df.mean()
```

```python
df.std()
```

```python
(df-df.mean())/df.std()
```

## Min-max scaling

```python
df.max()
```

```python
df.min()
```

```python
(df-df.min())/(df.max()-df.min())
```
